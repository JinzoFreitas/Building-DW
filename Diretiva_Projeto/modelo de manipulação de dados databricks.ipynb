```
file_location = "/FileStore/tables/dataSUS_part0000.csv"

file_type = "csv"

infer_schema = "false"
first_row_is_header = "true"
delimiter = ";"

df = spark.read.format(file_type) \
    .option("inferSchema", infer_schema) \
    .option("header", first_row_is_header) \
    .option("sep", delimiter) \
    .load(file_location)

display(df)
```

### Criando um arquivo parquet

```
df.write.format("parquet").option("header","true").save("dbfs:/FileStore/raw/opendatasus")
```

```
df.printSchema()
```
